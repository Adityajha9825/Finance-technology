```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

```

```python
sns.countplot(data=df,x='survived')
vs=df['survived'].value_counts('sex')
print(vs)
```

```python
sns.countplot(x="pclass",hue="sibsp",data=df,palette="rainbow")
vs=df['survived'].value_counts('pclass'),df['survived'].value_counts('sibsp')
print(vs)
```
